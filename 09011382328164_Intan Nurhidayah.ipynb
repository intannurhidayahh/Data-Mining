```python
import pandas as pd
import matplotlib.pyplot as plt

# Pilih hanya kolom numerik
numeric_cols = df.select_dtypes(include='number')

# Membuat histogram untuk setiap kolom numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(10, 6))
    plt.hist(numeric_cols[column].dropna(), bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Membuat box plot untuk setiap kolom numerik
for column in numeric_cols.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=numeric_cols[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Contoh data frame
data = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 6, 7, 8, 9],
    'D': [10, 20, 30, 40, 50]
})

# Menghitung matriks korelasi
corr_matrix = data.corr()

# Menyaring pasangan variabel dengan korelasi tinggi (misalnya > 0.8 atau < -0.8)
high_corr_pairs = [(i, j) for i in corr_matrix.columns for j in corr_matrix.columns
                   if i != j and abs(corr_matrix.loc[i, j]) > 0.8]

# Membuat scatter plot untuk setiap pasangan variabel dengan korelasi tinggi
for (x, y) in high_corr_pairs:
    sns.scatterplot(data=data, x=x, y=y)
    plt.title(f'Scatter plot antara {x} dan {y}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
    
```
